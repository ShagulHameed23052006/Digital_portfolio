Linear regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data: hours studied vs math scores
hours_studied = np.array([2, 4, 6, 8, 10]).reshape((-1, 1))
math_scores = np.array([75,80, 85, 90, 100])

# Create and fit the linear regression model
model = LinearRegression()
model.fit(hours_studied, math_scores)

# Predict math scores
hours_to_predict = np.array([15]).reshape((-1, 1))
predicted_score = model.predict(hours_to_predict)
print(f"Predicted math score for {hours_to_predict[0][0]} hours studied: {predicted_score[0]}")

# Plot the data and regression line
plt.scatter(hours_studied, math_scores, color='blue', label='Actual scores')
plt.plot(hours_studied, model.predict(hours_studied), color='red', label='Regression line')
plt.xlabel('Hours studied')
plt.ylabel('Math scores')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
