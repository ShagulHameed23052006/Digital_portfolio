Matplotlib

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 14, 19, 23, 15]

# Plotting
plt.scatter(x, y)

# Labels and title
plt.xlabel("X-axis (Time)")
plt.ylabel("Y-axis (Value)")
plt.title("Simple Scatter Chart")

# Show the chart
plt.show()
```
